### 1.Normalizing Data <br>
##### Categories: 0-building; 1-barren land; 2-trees; 3-grassland; 4-road; 5-water.


```python
zip_77019 = np.load('../input_data/zip_77019_4bands.npy')
# save image and rbg bands 
zip_77019_image = zip_77019[:,:,0:3]
# print shape 
np.shape(zip_77019_image)
```

```python
# plot image of 77019
plt.imshow(zip_77019_image)
```

```python
# max list to hold the maximum number in every element of the list
max_list = []
# loop through the image list
for i in zip_77019_image:
    # loop through each element and get the max value 
    for j in i:
        max_list.append(max(j))
```

```python
plt.plot(max_list)
plt.show()
```

```python
len(zip_77019_image[0][0])
```

```python
scaled_image = zip_77019_image*255
```

```python
## plot max values for the scaled image 
# max list to hold the maximum number in every element of the list
max_list_scaled = []
# loop through the image list
for i in scaled_image:
    # loop through each element and get the max value 
    for j in i:
        max_list_scaled.append(max(j))

# plot and check integrity 
plt.plot(max_list_scaled)
plt.show()
```
